## __Fundamental Principle of Counting__ 

It states that if there are "m" ways of doing something and "n" ways of doing another thing, then there are "m times n" ways of performing both actions.

__Example__
Suppose a person have 2 shirts (green and yellow) and 3 jeans (red, blue, purple) it implies that he have 6 options of chossing between 2 shirts and 3 jeans.
Using tree visualization, count the paths from root to leaves.

```python
# Example Data for Notebook
n_items = [1, 2, 3, 4, 5]
n = len(n_items)
r_values = list(range(1,1 + len(n_items)))
```

## __Permutations__

__Permutations__ are the each ARRANGEMENTS in pairs of r from n given items.
- Order of paris matters
- With and without replacement
- With replacement = pow(n,r)
- nPr = n!/(n-r)!

__Example__

Suppose we have unique list of items

- items = [a,b,c,d,e]
- n = 5 
- r = 3

Consider we have 3 chairs _ * _ * _ to fill from the given list of letters


__Permutations WITHOUT Replacement__


Without replacement _ * _ * _ can be filled in 5 ways times 4 ways times 3 ways (Fundamental Principal of Counting) which is mathematically nPr = n!/(n-r)!. 


__Example of Permutations WITH Replacement__

 With replacement _ * _ * _ can be filled in 5 ways times 5 ways times 5 ways (Fundamental Principal of Counting) which is mathematically pow(n,r)

__Point to Ponder__

When the items are not unique then divide nPr by the all repeating item's counts factorrial. For example for [a,b,c,c,b] would be nPr / (2! * 2!) or pow(n,r) / (2! * 2!)



### __Permutations Computation [Without Repetition]__

```python
# import math.perm is just for testing counts

# from math import perm

perm_dict = {}
perm_count = {}
for r in r_values:
    perm_dict[r] =  list(permutations(n_items, r=r))
    perm_count[r] = len(perm_dict[r]) 
    # assert(perm(n,r) == )
print(f'No. of permutations for each r = {perm_count}')
print(f'Exhaustive list of permutations is {perm_dict}')
```

### __Permutations Computation [With Repetition]__

```python
perm_with_rep_count = {}
perm_with_rep = {}

for r in r_values:
    perm_with_rep[r] = list(product(n_items, repeat=r))
    perm_with_rep_count[r] = len(perm_with_rep[r])
    assert(pow(n,r) == len(perm_with_rep[r]))

print(perm_with_rep_count)
print(perm_with_rep)
```

## __Combinations__

__Combinations__ are the each SELECTIONS in pairs of r from n given items 
- Order of pairs doesn't matter
- With and without replacement
- With replacement = r+n-1 C r
- nCr = n!/r! * (n-r)!



__Example__

Suppose we have unique list of items

- items = [a,b,c,d,e]
- n = 5 
- r = 3

Consider we have team of 3 (set not arrangement) _ * _ * _ to fill from the given list of letters

__Combinations WITHOUT Replacement__

Without replacement _ * _ * _ can be filled in 5 ways times 4 ways times 3 ways / 3! (for repetition due to selection) which is mathematically nCr = n!/(r!*(n-r)!).

__Combinations WITH Replacement__

Mathematically can be expressed as r+n-1 C r. Let's try to visualize it using a problem scenario

- __Problem__ Let us say there are five flavors of icecream: banana, chocolate, lemon, strawberry and vanilla. We can have three scoops. How many variations will there be?

- __Intuition__ Think about the ice cream being in boxes, we could say "move past the first box, then take 3 scoops, then move along 3 more boxes to the end" and we will have 3 scoops of chocolate! So it is like we are ordering a robot to get our ice cream, but it doesn't change anything, we still get what we want.

- __Solutoin__ We can write this down as __->ooo->->->__ (arrow means move, circle means scoop).In fact the three examples above can be written like this: {c, c, c} (3 scoops of chocolate): 	acccaaa. {b, l, v} (one each of banana, lemon and vanilla): 	caacaac. {b, v, v} (one of banana, two of vanilla): 	caaaacc

- __Problem Simplification__ OK, so instead of worrying about different flavors, we have a simpler question: "how many different ways can we arrange arrows and circles?"



Notice that there are always 3 circles (3 scoops of ice cream) and 4 arrows (we need to move 4 times to go from the 1st to 5th container). So (being general here) there are r + (n1) positions, and we want to choose r of them to have circles. This is like saying "We have r + (n1) pool balls and want to choose r of them". 

__Reference__: https://www.mathsisfun.com/combinatorics/combinations-permutations.html


### __Combinations Computation [Without Repetition]__

```python
# Just for testing of counts

from math import comb
comb_dict = {}
comb_count = {}

for r in r_values:
    comb_dict[r] = list(combinations(n_items, r=r))
    comb_count[r] = len(comb_dict[r])
    assert(comb(n,r) == comb_count[r]) 
    
print(f'No. of combinations for each r = {comb_count}')
print(f'Exhaustive list of combinations is {comb_dict}')
```

### __Combinations Computation [With Repetition]__

```python

comb_with_rep_count = {}
comb_with_rep = {}

for r in r_values:
    comb_with_rep[r] = list(combinations_with_replacement(n_items, r))
    comb_with_rep_count[r] = len(comb_with_rep[r])
    assert(comb(n-1+r,r) == comb_with_rep_count[r]) 
    
print(comb_with_rep_count)
print(comb_with_rep)
```
